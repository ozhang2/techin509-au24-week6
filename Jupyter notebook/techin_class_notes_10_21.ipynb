## üñä Practice 4: Normalization

```python
# add your code below
def normalize(l):
    # set variable for sum of list
    sum_of_list = sum(l)
    # make an empty list for the normalized list
    normalized_l = []

    # for loop to go through the index from the list
    for i in l:
        # equation for normalizing - float numbers
        normalized_i = float(i) / float(sum_of_list)
        # append normalized i to the new list
        normalized_l.append(normalized_i)

    # return the new list
    return normalized_l

# test your code using the examples below
l1 = [1, 2, 3]
normalized_l1 = normalize(l1)
print(normalized_l1)

l2 = list(range(20))
normalized_l2 = normalize(l2)
print(normalized_l2)
```

# Pure functions and side effects

A pure function is a function that always returns the same output given a specific input.

A pure function does not change any value or state outside of its scope and it doesn't depend on any value outside of its scope. It only depends on the input given to the function and does not produce any side effects.

```python
# pure function
def add(x,y):
    return x+ y

add(1,2)
```

```python
# side effects
default_val = 0
def add_to_default(x):
  return x + default_val

add_to_default(1)

default_val = 5
add_to_default(1)
```

# Handling Errors

```python
number_str = input("give me a number:")

try:
  ans = int(number_str) * 2
  print(ans)
except ValueError:
  print("please input a valid number")
```

```python
def divide(x,y):
  try:
      ans = x/y
      return ans
  except ZeroDivisionError:
      print('cannot divide by zero')
  else:
      print("there is no error")
      print(ans)
  finally:
      print("let's clean up the mess... whether there is error or not")

divide(4,2)
```

```python
# important when opening, reading, writing and closing files

fd = open('/tmd/a.txt')
fd.read()
fd.write("hello")

try:
  fd.read()
  fd.write("hello")
except:
  print("something went wrong")
finally:
  fd.close()

# OR use 'with' - it's the same thing

with open('/tmd/a.txt') as fd:
  fd.read()
  fd.write("hello")
```

## üñä Error Handling Practice

```python
# ask the user to input a valid number
# if the user gives an invalid number, ask again
# ask the user for up to 3 times, then exit the program and print "Goodbye"

tries = True
count = 0

while tries:
  number_str = input("give me a number:")
  try:
    ans = int(number_str) * 2
    print(ans)
    tries = False
  except ValueError:
    count += 1
    if count == 3 and tries:
      print("Goodbye")
      tries = False
    else:
      print("please input a valid number")
```

```python
# class solution
retry_cnt = 0

while retry_cnt < 3:
  number_str = input("give me a number:")
  try:
    ans = int(number_str) * 2
    print(ans)
  except ValueError:
    print("please input a valid number")
    retry_cnt += 1
```

```python
# another way class solution

retry_cnt = 1

while True:
  if retry_cnt == 4:
      print("Goodbye")
      break
  number_str = input("give me a number:")
  try:
    ans = int(number_str) * 2
    print(ans)
    break
  except ValueError:
    print("[%s/3] please input a valid number"%retry_cnt)
    retry_cnt += 1
```

```python
import requests
from pypdf import PdfReader
from wordcloud import WordCloud
import matplotlib.pyplot as plt

url = "https://www.federalreserve.gov/mediacenter/files/FOMCpresconf20230920.pdf"

# first download a PDF from a press conference (you can use any PDF you would like)
res = requests.get(url)
open('speech.pdf', 'wb').write(res.content)

# using the packages we imported, we can extract the text (for certain PDFs)
reader = PdfReader("speech.pdf")
page = reader.pages[0]
text = page.extract_text()

# check the content to make sure
# Check the content - first 500 characters
print(text[:500])
```

```python
wordcloud = WordCloud().generate(text)
plt.imshow(wordcloud, interpolation = 'bilinear')
plt.axis('off')

# most commonly mentioned words in this text
```
